# Business Problem Description

Here Building a machine learning model to predict how many passengers survied in Titanic massacre

# EDA

```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df.head()
```

```python
inputs = df.drop('Survived',axis='columns',)
inputs
```

```python
target = df['Survived']
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_PassengerId = LabelEncoder()
le_Pclass = LabelEncoder()
le_Name = LabelEncoder()
le_Sex = LabelEncoder()
le_Age = LabelEncoder()
le_SibSp = LabelEncoder()
le_Parch = LabelEncoder()
le_Ticket = LabelEncoder()
le_Fare = LabelEncoder()
le_Cabin = LabelEncoder()
le_Embarked = LabelEncoder()
```

```python
inputs['PassengerId_n'] = le_PassengerId.fit_transform(inputs['PassengerId'])
inputs['Pclass_n'] = le_Pclass.fit_transform(inputs['Pclass'])
inputs['Name_n'] = le_Name.fit_transform(inputs['Name'])
inputs['Sex_n'] = le_Sex.fit_transform(inputs['Sex'])
inputs['Age_n'] = le_Age.fit_transform(inputs['Age'])
inputs['SibSp_n'] = le_SibSp.fit_transform(inputs['SibSp'])
inputs['Parch_n'] = le_Parch.fit_transform(inputs['Parch'])
inputs['Ticket_n'] = le_Ticket.fit_transform(inputs['Ticket'])
inputs['Fare_n'] = le_Fare.fit_transform(inputs['Fare'])
inputs['Cabin_n'] = le_Cabin.fit_transform(inputs['Cabin'])
inputs['Embarked_n'] = le_Embarked.fit_transform(inputs['Embarked'])
inputs
```

```python
inputs_n = inputs.drop(['PassengerId','Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],axis='columns')
inputs_n
```

# Testing 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(inputs_n,target,test_size=0.25)
```

```python
print(X_train.head(),'\n\n',X_test.head())
print('\n',X_train.shape,'\n\n',X_test.shape)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model = tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

# Accuracy of the Model

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=model.predict(X_test)
print(y_pred)
```

```python
accuracy = accuracy_score(y_test,y_pred)
```

```python
print("Accuracy of the model:\n\n as per the data is \n\n",accuracy)
```
